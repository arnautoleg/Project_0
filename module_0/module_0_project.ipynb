```python
import numpy as np                     # импорт библиотеки numpay

def game_core(number):
    '''Устанавливаем верхнюю и нижнюю границы диапазона, выбирая середину в качестве предсказанного числа.
       С помощью while этот диапазон сужаем, регистрируя каждую попытку.
       Функция принимает загаданное число и возвращает число попыток'''
    count = 1 
    predict = 50                    # выбираем середину в качестве предсказанного числа
    low_limit, hight_limit = 1, 101 # устанавливаем верхнюю и нижнюю границы диапазона
    
    while number != predict: #  выход из цикла, если угадали
        count+=1
        if predict > number: hight_limit = predict           
        else: low_limit = predict   
        predict = low_limit + (hight_limit - low_limit)//2 # предсказанное число для следующей итерации расчитывается исходя из границ
    return count 


def score_game(game_core):
    '''Запускаем алгоритм 1000 раз, чтобы узнать среднее число попыток для угадывания числа'''
    count_ls = []
    np.random.seed(1)                   #   фиксируем RANDOM SEED, чтобы эксперимент был воспроизводим
    random_array = np.random.randint(1,101, size=(1000)) # загадали число от 1 до 100 тысячу раз и сохраняем в массив
    
    for number in random_array:
        count_ls.append(game_core(number)) # заполняем пустой список количеством попыток для отдельно взятых значений из random_array
    
    mean_attempts = int(np.mean(count_ls)) # расчитываем среднее число попыток
    print(f"Ваш алгоритм угадывает число в среднем за {mean_attempts} попыток") 
    return(mean_attempts)


score_game(game_core)
```
